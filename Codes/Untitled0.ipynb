```python
import warnings
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
# set options to be headless, ..
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
import time
import re
from tqdm import tqdm
```

```python
req = requests.get('https://www.koreagoldx.co.kr/')
html = req.text
soup = BeautifulSoup(html, 'lxml')
```

```python
# open it, go to a website, and get results
driver = webdriver.Chrome('chromedriver',options=options)
URL = 'https://www.koreagoldx.co.kr/'
driver.get(url=URL)
```

```python
gold = {'일자': [], '내가 살 때(VAT 포함)': [], '내가 팔 때': []}
silver = {'일자': [], '내가 살 때(VAT 포함)': [], '내가 팔 때': []}
```

```python
date = driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div[1]/div[2]/div/table[1]/caption/span').text
gold_s = driver.find_element_by_xpath('//*[@id="spure"]/span').text
gold_p = driver.find_element_by_xpath('//*[@id="ppure"]/span').text
siver_s = driver.find_element_by_xpath('//*[@id="ssilver"]/span').text
silver_p = driver.find_element_by_xpath('//*[@id="psilver"]/span').text
```

```python
gold['일자'].append(date)
gold['내가 살 때(VAT 포함)'].append(gold_s)
gold['내가 팔 때'].append(gold_p)
silver['일자'].append(date)
silver['내가 살 때(VAT 포함)'].append(siver_s)
silver['내가 팔 때'].append(silver_p)
```
