### 데이터 정제

```python
import pandas as pd

dic = pd.read_excel('dictionary.xlsx', header=None)
```

```python
dic.columns = ['word']
```

```python
dic['word'] = dic['word'].str.lower() # 소문자로 통일
```

```python
dictionary = pd.DataFrame({'term': dic['word'], 'a2': None, 'a3': None, 'a4': 0, 'tag': 'NNG', 'a6': '*', 'batchim': 'F', 'terms2': dic['word'], 'a9': '*', 'a10': '*', 'a11': '*', 'a12': '*', 'a13': '*'})
# 단어 비용이 낮을 수록 우선순위가 높아진다.
# 따라서 사용자 사전의 단어들이 우선적으로 나타나도록 단어 비용을 0으로 통일하였으며, 형태소를 편의 상 'NNG'로 통일하였다.
```
